```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv(r"mnist_train.csv")
df1
```

```python
X_train=df1.iloc[:,1:]
y_train=df1.iloc[:, 0]
```

```python
df2=pd.read_csv(r"mnist_test.csv")
df2
```

```python
X_test = df2.iloc[:,1:]
y_test = df2.iloc[:, 0]
```

```python
from sklearn.preprocessing import Normalizer
```

```python
norm = Normalizer(norm='max')
```

```python
norm.fit(X_train)
```

```python
X_train = norm.transform(X_train)
```

```python
X_test = norm.transform(X_test)
```

```python
X_train
```

```python
#MLPClassifier
from sklearn.neural_network import MLPClassifier
```

```python
clf1 = MLPClassifier(hidden_layer_sizes=(10,10), validation_fraction=0.1, max_iter=500, verbose=1, learning_rate='adaptive', early_stopping=True, solver='adam', activation='relu')
```

```python
clf1.fit(X_train, y_train)
```

```python
y_pred = clf1.predict(X_test)
```

```python
from sklearn.metrics import f1_score, accuracy_score, classification_report, roc_auc_score, roc_curve, auc
```

```python
acc_score = accuracy_score(y_test, y_pred)
```

```python
acc_score
```

```python
f1Score = f1_score(y_test, y_pred, average='macro')
```

```python
f1Score
```
